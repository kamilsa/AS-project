```python
tizen_all = pd.read_csv('results/tropico.csv')
res = [tizen_all[tizen_all.percentage == 10].method.reset_index()]
for i in range(10,100, 10):
    res.append(tizen_all[tizen_all.percentage == i].acfp.reset_index())
res = pd.concat(res, axis=1)
del res['index']
print(res.round(4).to_latex())
```

```python
x = range(10,100, 10)

column = 'acfp'
ffn_sigmoid = tizen_all[tizen_all.method == 'ffn_sigmoid'][column].values
ffn_custom = tizen_all[tizen_all.method == 'ffn_custom'][column].values
rnn_sigmoid = tizen_all[tizen_all.method == 'rnn_sigmoid'][column].values
rnn_custom = tizen_all[tizen_all.method == 'rnn_custom'][column].values
rnn_lstm = tizen_all[tizen_all.method == 'rnn_lstm'][column].values
logistic = tizen_all[tizen_all.method == 'logistic'][column].values
go_concave = tizen_all[tizen_all.method == 'go_concave'][column].values
go_s_shaped = tizen_all[tizen_all.method == 'go_s_shaped'][column].values
weibull = tizen_all[tizen_all.method == 'weibull'][column].values
```

```python
colors = ('b', 'g', 'r', 'c', 'm', 'y', 'k')
plt.plot(x, ffn_sigmoid, 'r-', \
         x, ffn_custom, 'b-', \
         x, rnn_sigmoid, 'r--', \
         x, rnn_custom, 'b--', \
         x, rnn_lstm, 'k--', \
         x, logistic, 'g:', \
         x, go_concave, 'y:', \
         x, go_s_shaped, 'c:', \
         x, weibull, 'r:', \
        linewidth=3, alpha=0.6)
# plt.ylim(0,0.2)
# plt.yscale('log')
plt.legend(('ffn sigmoid', 'ffn custom', 'rnn sigmoid', 'rnn custom', 'rnn lstm', \
            'logistic', 'go concave', 'go s-shaped', 'weibull'), loc=1)
plt.show()
```
